```python
from collections import deque
```

```python
graph = {
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':['F'],
    'F':[]
}
```

```python
def bfs(start):
    visited = []
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.append(node)
            queue.extend(graph[node])

print("BFS Traversal:")
bfs('A')
```

```python
graph = {
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':['F'],
    'F':[]
}

visited=[]
```

```python
def dfs(node):
    if node not in visited:
        print(node, end=" ")
        visited.append(node)
        for neighbour in graph [node]:
            dfs(neighbour)

print("DFS Traversal:")
dfs('A')
```
