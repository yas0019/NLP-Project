```python
from textblob import TextBlob
```

```python
#Sample reviews
reviews= [
    "I love this product, it is amazing!",
    "This is the worst purchase I never made.",
    "The product is okay, not great."
]
```

```python
#statement analysis
for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    if polarity>0:
        sentiment = "positive"
    elif polarity<0:
        sentiment = "Negative"
    else:
        sentiment = "Natural"
    print(f"Review: {review}\nSentiment:{sentiment}\nPolarity:{polarity}\n")
```
