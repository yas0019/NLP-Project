```python
import nltk
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.probability import FreqDist
```

```python
text = "Language modeling is an important part of Natural Language Processing"
```

```python
tokens = word_tokenize(text.lower())
```

```python
#General N grams (n=2)
bigrams = list(ngrams(tokens,2))
```

```python
freq= FreqDist(bigrams)
```

```python
print("Bigram Frequency:")
for k, v in freq.items():
    print(k, ":", v)
```

```python
print("\nMost Common Bigrms:")
print(freq.most_common(5))
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.probability import FreqDist
```

```python
text = "Natural language processing enables to understand human language."
```

```python
tokens = word_tokenize(text.lower())
```

```python
unigrams= tokens 
bigrams = list (ngrams(tokens, 2))
trigrams = list(ngrams(tokens,3))
```

```python
print("Unigrams:", unigrams)
print("Bigrams :", bigrams )
print("Tigrams :", trigrams )
```

```python
print("\Most Common Unigrams:")
print(FreqDist(unigrams).most_common(5))
```

```python
print("Most Common Bigram:")
print(FreqDist(bigrams).most_common(5))
```

```python
print("Most Common Trigram:")
print(FreqDist(trigrams).most_common(5))
```
